```python
# Import dependencies
import pandas as pd
```

```python
# Read csv into dataframe
aqi = pd.read_csv('resources/Pollution_day_2001-2016.csv')
aqi.head(5)
```

```python
# Filter data by California
aqi_cal = aqi.loc[aqi['State'] == 'California']
```

```python
# Print county names
aqi_cal['County'].unique().tolist()
```

```python
# Select only years 2009 to 2012
aqi_cal_2012 = aqi_cal.loc[(aqi_cal['Date Local'].str.contains('2012'))]
aqi_cal_2011 = aqi_cal.loc[(aqi_cal['Date Local'].str.contains('2011'))]
aqi_cal_2010 = aqi_cal.loc[(aqi_cal['Date Local'].str.contains('2010'))]
aqi_cal_2009 = aqi_cal.loc[(aqi_cal['Date Local'].str.contains('2009'))]
```

```python
# Merge data into one dataframe
aqi_data = aqi_cal_2012.merge(aqi_cal_2011, how='outer')
aqi_data = aqi_data.merge(aqi_cal_2010, how='outer')
aqi_data = aqi_data.merge(aqi_cal_2009, how='outer')
```

```python
# Save dataframe to csv file
aqi_data.to_csv('resources/Cleaned_folderaqi_data.csv', index=False)
```

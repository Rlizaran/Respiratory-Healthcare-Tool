```python
import pandas as pd
```

```python
aqi = pd.read_csv('aqi_data.csv')
asthma = pd.read_csv('asthma_cleaned.csv')
```

```python
asthma.columns.tolist()
```

```python
asthma_clean = asthma.drop(['County Fips code', 'Lat','Lon','ZIP code'], axis=1).copy()
```

```python
asthma_clean
```

```python
asth = asthma_clean.groupby(['Year','Age Group',pd.Grouper(key='County')]).mean()
```

```python
asthma_to_merg = asth.reset_index()
```

```python
aqi = aqi.drop('State',axis=1)
```

```python
len(asthma_clean['County'].unique().tolist())
```

```python
aqi['Year'] = aqi['Date Local'].str.split('-').str[0]
```

```python
aqi['Year'] = pd.to_numeric(aqi['Year'])
```

```python
aqi = aqi.drop('Date Local', axis = 1)
```

```python
aqi.columns
```

```python
aqi_grouped = aqi.groupby('Year')
```

```python
aqi_grouper = aqi.groupby(['Year',pd.Grouper(key='County')]).agg(['mean','max','min'])
```

```python
aqi_grouper
```

```python
aqi_cleaned = aqi_grouper.reset_index()
```

```python
aqi_cleaned
```

```python
aqi_cleaned['County'] = aqi_cleaned['County'].str.lower().str.capitalize()
```

```python
asthma_to_merg['Year'].dtype
```

```python
merged = asthma_to_merg.merge(aqi_cleaned, how='inner', on=['County','Year'])
```

```python
print(f'rows: {len(merged)}, columns: {len(merged.columns.tolist())}')
```

```python
merged
```

```python
merged.to_csv('aqi_asthma.csv', index=False)
```
